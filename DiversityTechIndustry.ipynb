```python
def discover_feature_list(column):
    '''
    INPUT 
        column - a dataframe series
        
    OUTPUT
        return - a list of possible values to a given column
    '''
    attr_set = set()
    for row in column:
        if type(row) == str:
            attr_itens = row.split(';')
            for item in attr_itens:
                attr_set.add(item)
                
    return list(attr_set)


def findings_by_minority(years_df, minority, years_minority, label, func):
    '''
    INPUT 
        years_df - a dict with dataframes for each one of the study years;
        minority - a dict with strings representing the column name of minority study for each one of 
                   the study years;
        years_minority - a dict with each possible values for a given column setted above;
        (The values can assume differente string to say the same think accros the years);
        label - a string representing name used to group the "minorities";
        func - a callback function to return the value represented by some calculation made in this function;
    OUTPUT
        return - a dataframe containing values represented some statistic calculated in "func" by each possible 
                 values setted in years_minority;
    '''
    years = list(years_df.keys())
    minorities = years_minority.keys()
    data = []
    
    for item in minorities:
        row = []
        
        for year in years:
            num = np.nan
            df = years_df[year]
            feature_name = minority[year]
            
            if years_minority[item][year]:
                num = func(df[df[feature_name].str.contains(years_minority[item][year], na=False)], year, feature_name)
                            
            row.append(num)
            
        last_year = years[-1]
        row.insert(0, years_minority[item][last_year])
        data.append(row)
        
    years.insert(0, label)
    new_df = pd.DataFrame(data, columns=years)
    new_df = new_df.set_index(label)
    new_df.sort_values(by=years[-1], ascending=False)
    return new_df


def data_crossing(df, country, factor1, factor_list, factor2, percentage=True):
    '''
    INPUT 
        df - a dataframe representing the result of a specific query;
        country - a string representing the country of study;
        factor1 - a string representing the study column name;
        factor_list - a list with each possible value of study for a given column setted above;
        factor2 - a string representing the second study colum name;
        percentage - a boolean with true if return should be calculated using proportion instead 
                     of absolute numbers; (true by default)
    OUTPUT
        return - a dataframe containing values for a given data crossing;
                 Example: Ethnicities X Job Satisfaction
    '''
    country_df = df.query('Country == "{}"'.format(country))
    columns = discover_feature_list(df[factor2])
    data = []
    
    for item in factor_list:
        minotity_df = country_df[country_df[factor1].str.contains(item, na=False)]
        sat_df = (minotity_df[factor2].value_counts()/minotity_df[factor2].count()).reset_index()
        
        if not percentage:
            sat_df = minotity_df[factor2].value_counts().reset_index()
            
        sat_df.set_index('index', inplace=True)
        
        row = [item]
        for column in columns:
            row.append(float(sat_df.loc[column]))
            
        data.append(row)
        
    columns.insert(0, factor1)
    new_df = pd.DataFrame(data, columns=columns)
    new_df.set_index(factor1, inplace=True)
    
    return new_df


def filter_data_by_column(df, column_name, column_label):
    '''
    INPUT 
        df - a dataframe holding the CousinEducation column
        title - string the title of your plot
        axis - axis object
        plot - bool providing whether or not you want a plot back
        
    OUTPUT
        study_df - a dataframe with the count of how many individuals
        Displays a plot of pretty things related to the CousinEducation column.
    '''
    study = df[column_name].value_counts().reset_index()
    study.renEduOtherame(columns={'index': column_name, column_name: column_label}, inplace=True)

    return study


def merge_indexes(df, relations):
    
    for relation in relations:
        origin_index = relation['origin']
        destiny_index = relation['destiny']
        keys = df.loc[origin_index].keys()
        for key in keys:
            if df.isnull().loc[destiny_index, key]:
                df.loc[destiny_index][key] = df.loc[origin_index][key]
        
        df = df.drop(index = origin_index)

    return df
                  
    
def remove_false_positives(df, country, minority, minority_list, column_label):
    country_query = 'Country == "{}"'.format(country)
    satisfaction_query = 'JobSat == "Very satisfied"'
    jobseek_query = 'JobSeek == "I am actively looking for a job"'
    data = []     
     
    query_df = df.query('({} & ({}) & {})'.format(country_query, employment_query, satisfaction_query))    
    for by in minority_list:    
        minority_df = query_df[query_df[minority].str.contains(by, na=False)]
        false_positive = minority_df.query(jobseek_query).shape[0]
        data.append([by, false_positive/minority_df.shape[0]])
        
    new_df = pd.DataFrame(data, columns=[minority, column_label])
    new_df.set_index(minority, inplace=True)
    
    return new_df


employment_query = 'Employment == "Employed full-time" | Employment == "Employed part-time" | Employment == "Independent contractor, freelancer, or self-employed"'                       
```

#### Question 1: How diverse are tech workforce in the US?
In order to understand more about how diversy are the US tech workforce, it's necessary understand distribution by ethnicity and gender.

```python
print('Total sample of 2018 is {}'.format(df18.query('Country == "United States"').shape[0]))
print('Total sample of 2019 is {}'.format(df19.query('Country == "United States"').shape[0]))
print('Total sample of 2020 is {}'.format(df20.query('Country == "United States"').shape[0]))
```

```python
def get_distribution_by_years(df, year, minority):
    '''
    Callback function to findings_by_minority
    
    INPUT 
        df - a dataframe of specific year, result of a previous query
        year - year of the dataframe
        minority - type of minority (Ethnicity | Gender)
        
    OUTPUT
        return - proportion of a specific group (Ethnicity | Gender), in the US for a specific year
     '''
    
    query_by_country = 'Country == "United States"'
    
    if year == 2018:
        df_18 = df18.query(query_by_country)
        return df.shape[0]/df_18[minority].count()

    elif year == 2019:
        df_19 = df19.query(query_by_country)
        return df.shape[0]/df_19[minority].count()
    
    elif year == 2020:
        df_20 = df20.query(query_by_country)
        return df.shape[0]/df_20[minority].count()
    
    return np.nan
```

According to the above data crossing, it’s possible to conclude that minorities in the tech workforce in the US are Hispanics or Latinos, Middle eastern and Black or African descent by ethny and Women and Non-binary or non-conforming by gender.

How satisfied are minorities ethnicities among their peers?

Stackoverflow asked developers which their job seeking status was in 2019 and 2020. So, it’s possible to find some false negatives, i.e, developers who said they  were very satisfied with their jobs and were actively looking for a job.

```python
# Extracting the false positives related to people who felt very 
# satisfied and were actively looking for a new job by ethnicity

df_20 = remove_false_positives(df20, 'United States', 'Ethnicity', ethnicity_list20, '2020')
df_19 = remove_false_positives(df19, 'United States', 'Ethnicity', ethnicity_list19, '2019')

relations = [{'origin': 'Hispanic or Latino/Latina', 'destiny': 'Hispanic or Latino/a/x'}]
ethny_false = merge_indexes(pd.concat([df_19, df_20], axis=1, sort=False), relations)
ethny_false
```

```python
# Extracting the false positives related to people who felt very 
# satisfied and were actively looking for a new job by gender

df_20 = remove_false_positives(df20, 'United States', 'Gender', gender_list20, '2020')
df_19 = remove_false_positives(df19, 'United States', 'Gender', gender_list19, '2019')

gender_false = merge_indexes(pd.concat([df_19, df_20], axis=1, sort=False), [])
gender_false
```

#### Question 3: Do minorities have been compensated equally in the US tech industry?

According to Forbes, salary is correlated to job satisfaction. So, it’s important to understand the difference in salaries across the last years.

```python
def get_salary_by_years(df, year, minority):
    '''
    Callback function to findings_by_minority
    It's important stands out the high variability of annual salary.
    So was used the median to better representation of annual salary of the samples.
    
    INPUT
        df - a dataframe of specific year, result of a previous query
        year - year of the dataframe
        minority - type of minority (Ethnicity | Gender)
        
    OUTPUT
        return - annual salary for a specific group (Ethnicity | Gender), in the US for a specific year
     '''
        
    salary_list = {2018: 'ConvertedSalary', 2019: 'ConvertedComp', 2020: 'ConvertedComp'}
    salary_feature_name = salary_list[year]
    return df.median()[salary_feature_name]
```

```python
query_by_country = 'Country == "United States" & ({})'.format(employment_query)
df_18 = df18.query(query_by_country)
df_19 = df19.query(query_by_country)
df_20 = df20.query(query_by_country)

df_list = {2018: df_18, 2019: df_19, 2020: df_20}
minority = {2018: 'RaceEthnicity', 2019: 'Ethnicity', 2020: 'Ethnicity'}
ethnicities = {
    'White': {2018: 'White or of European descent', 2019: 'White or of European descent', 2020: 'White or of European descent'},
    'Middle': {2018: 'Middle Eastern', 2019: 'Middle Eastern', 2020: 'Middle Eastern'},
    'Hispanic': {2018: 'Hispanic or Latino/Latina', 2019: 'Hispanic or Latino/Latina', 2020: 'Hispanic or Latino/a/x'},
    'Black': {2018: 'Black or of African descent', 2019: 'Black or of African descent', 2020: 'Black or of African descent'}
}

ethnicities_salary = findings_by_minority(df_list, minority, ethnicities, 'Ethnicity', get_salary_by_years)
ethnicities_salary
```

```python
ethnicities_salary.plot(kind='bar')
```

```python
query_by_country = 'Country == "United States"'
df_18 = df18.query(query_by_country)
df_19 = df19.query(query_by_country)
df_20 = df20.query(query_by_country)

df_18 = df18.query('Country == "United States" & ({})'.format(employment_query))
df_19 = df19.query('Country == "United States" & ({})'.format(employment_query))
df_20 = df20.query('Country == "United States" & ({})'.format(employment_query))
df_list = {2018: df_18, 2019: df_19, 2020: df_20}
minority = {2018: 'Gender', 2019: 'Gender', 2020: 'Gender'}
genders = {
    'Man': {2018: 'Male', 2019: 'Man', 2020: 'Man'},
    'Woman': {2018: 'Female', 2019: 'Woman', 2020: 'Woman'},
    'Non-binary': {2018: 'Non-binary, genderqueer, or gender non-conforming', 2019: 'Non-binary, genderqueer, or gender non-conforming', 2020: 'Non-binary, genderqueer, or gender non-conforming'},
}

gender_salary = findings_by_minority(df_list, minority, genders, 'Gender', get_salary_by_years)
gender_salary
```

#### Question 4: The US tech industry has been absorbing the minorities's labor?

One of the questions in Stackoverflow’s survey was related to the current employment status of the respondent.

In order to understand better if the tech industry has been absorbing the minorities’s labor, we need to understand the unemployment rate, i.e, people that are not employed, but are looking for work.

```python
def get_umployment_rate(df, year, minority):
    unemployment = df.query('Employment == "Not employed, but looking for work"')
    return unemployment.shape[0]/df.shape[0]
```
